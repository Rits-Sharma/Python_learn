```python
import numpy as np
```

**Numpy Indexing and Slicing of Vectors**

```python
arr = np.arange(11, 21)
arr
```

```python
arr[9]
```

```python
arr[1:5]
```

```python
arr[3:]
```

```python
arr[3::4]
```

**Numpy Indexing and Slicing of Matrix**

```python
arr = np.arange(1, 31).reshape(6, 5)
arr
```

```python
arr[1]
```

```python
arr[0, 4]
```

```python
arr[0:2, 1:3]
```

```python
slice = arr[3:6, 3:5]
slice
```

**Boolean Indexing**

```python
arr = np.arange(11, 21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
