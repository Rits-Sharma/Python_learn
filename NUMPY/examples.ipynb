**Valid Sudoku**

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
s.sum()
```

```python
b = np.sum(s, axis=1)               # rows axis = 1, column axis = 0
b[b == 45]
```

```python
b == 45
```

```python
c = np.sum(s, axis=0)
for i in c:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("For Columns -> Sudoku is valid")
```

```python
# check 3 X 3 block
s[0:3, 0:3]
```

```python
s[3:6, 3:6]
```

```python
for i in range(0, 9, 3):
    for j in range(0, 9, 3):
        n = s[i:i+3, j:j+3]
        print(n.sum())
```

**General Questions**

```python
# Columns: [Age, Maths.marks, Science.marks]
data = np.array([
    [18, 85, 92],
    [19, 75, 88],
    [17, 76, 95],
    [17, 65, 65],
    [20, 90, 93]
])
```

```python
# Get the shape of the matrix
data.shape
```

```python
# Find the average of marks of each student
np.mean(data[:, 0])
```

```python
# Extract the math marks of all students
data[:, 1]
```

```python
# Find the highest science marks
np.max(data[:, 2])
```

```python
# Get the details of students who scored more than 90 in math
data[data[:, 1] > 90]
```

```python
# Increase math marks of all students by 5
data[:,1] + 5
```

```python
# Calculate the average marks in each subject (column-wise mean)
np.mean(data[:, 1:], axis=0)
```

```python
# Get data of students who scored at least 80 in both subjects
data[(data[:,1] >= 80) & (data[:, 2] >= 80)]
```

```python
# Replace all science marks < 75 with 0
data[:, 2][data[:, 2] < 75] = 0
data
```
