__ARRAY OPERATORS__

```python
import numpy as np
```

**Arithmetic operations**

```python
a1 = np.array([1, 2, 3, 4, 5])
a2 = np.array([6, 7, 8, 9, 10])
```

```python
a1 + a2
```

```python
a1 * a2
```

```python
a1 / a2
```

```python
a1 ** a2
```

**Broadcasting**

```python
l = [10, 20, 30, 40]
arr = np.array(l)
```

```python
arr + 10
```

```python
arr2 = np.arange(1, 26).reshape(5, 5)
arr2
```

```python
arr2 = arr2 * 10
arr2
```

**Deep and Shallow Copy**

```python
a = np.arange(1, 21)
```

```python
slice = a[:5]
slice = slice * 10
slice
```

```python
a
```

```python
a[4:6]
```

```python
a
b = a
```

```python
b[0] = 99
b
```

```python
a
```

```python
c = a.copy()
c[9] = 55
c
```

```python
a
```

**Stacking Array**

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
```

```python
np.vstack((a, b))
```

```python
np.hstack((a, b))
```

```python
np.column_stack((a, b))
```

```python
c = np.arange(16).reshape(4, 4)
c
```

**Splitting Array**

```python
np.split(c, 2)
```

```python
np.vsplit(c, 4)
```
